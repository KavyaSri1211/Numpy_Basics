**Reshaping of Arrays**

```python
#Reshaping
#converting 1D  array to 2D array

arr = np.arange(10)
arr.reshape(2, -1)  # Setting to -1 automatically decides the number of cols


```

```python
#reshaping into 3x3 

grid = np.arange(1, 10).reshape((3, 3))
print(grid)
```

**Zeros,Ones and Full**

```python
#Zeros,Ones and Full

# 3 x 4 ndarray full of zeros
# np.zeros(shape)
x = np.zeros((3,4))
print(x)

# a 3 x 2 ndarray full of ones
# np.ones(shape)
y = np.ones((3,2))
print(y)

# 2 x 3 ndarray full of fives
# np.full(shape, constant value)
z = np.full((2,3), 5)
print(z)
```

**Array Slicing - 1D**

```python
#syntax:  arr[start:stop:step]

#(:) It is called as slice notation used to access the subarrays

x = np.arange(10)
x
```

```python
x[:5]  # first five elements
```

```python
x[:5]  # first five elements
```

```python
x[4:7]  # middle sub-array
```

```python
x[::2]  # every other element
```

```python
x[::2]  # every other element
```

```python
x[::-1]  # all elements, reversed
```

```python
x[5::-2]  # reversed every other from index 5
```

**Array Concatenation and Splitting**

```python
#concatenation of arrays
x = np.array([1, 2, 3])
y = np.array([3, 2, 1])
np.concatenate([x, y])
```

```python
#splitting of arrays
x = [1, 2, 3, 99, 99, 3, 2, 1]
x1, x2, x3 = np.split(x, [3, 5])
print(x1, x2, x3)
```

**Sorting**

```python
x = np.array([2, 1, 4, 3, 5])
np.sort(x)
```

```python
#argsort returns  the indices of the sorted elements
x = np.array([2, 1, 4, 3, 5])
i = np.argsort(x)
print(i)
```
